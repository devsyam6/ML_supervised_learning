```python
import mahotas as mh
import numpy as np
"""Reading the image as NUMPY ARRAY"""
image=np.array (mh.imread('/home/jimbrootan/Desktop/ML_data/garden.jpeg'),dtype=np.float64)/255
'''converted image diamentions into the touple'''

diamentions=tuple(image.shape)
print(diamentions)


width,height,depth=diamentions
print(width,height,depth)


image_flattned=np.reshape(image,(width*height,depth))
print("image shape: ",image_flattned)

```

```python
"""Applying K menas for the extraction"""

from sklearn.cluster import KMeans
from sklearn.utils import shuffle

image_sample=shuffle(image_flattned,random_state=11)[:1000]
image_sample.shape

'''Compression of image using 64bit pixels'''

model=KMeans(n_clusters=64,random_state=20)
model.fit(image_sample)

cluster_assignments=model.predict(image_flattned)
compressed_palette=model.cluster_centers_
compressed_palette.shape
```

```python
'''Compressed image'''

compressed_image=np.zeros((width,height,compressed_palette.shape[1]))
print(compressed_image.shape)

label_index=0


for i in range(width):
    for j in range(height):
        compressed_image[i][j]=compressed_palette[cluster_assignments[label_index]]
        label_index+=1
```

```python
"""PLOTTING THE COMPRESSED IMAGE USING MATPLOTLIB.pyplot"""

import matplotlib.pyplot as plt


plt.subplot(122)
plt.title('Original Image')
plt.imshow(image)
plt.axis('off')
plt.subplot(121)
plt.title('Compressed image')
plt.imshow( compressed_image)
plt.show()
```
