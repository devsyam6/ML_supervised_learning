```python
%pylab inline

M=np.zeros((250,150))
M.shape

M[:31,:]=1
M[:,60:91]=1
M[-31:,:60]=1
M[150:,:31]=1
```

```python
"""adding some noice   """

M_noise=np.asmatrix(np.random.uniform(low=0,high=0.7,size=(250,150)))
M_noise.shape
M_noise=M+M_noise
M_noise
```

```python
"""PLOTTING DATA and using SVD"""


plt.subplot(1,5,1)
U,s,V=svd(M_noise)

plt.title("input image")
plt.imshow(M_noise)
plt.show()


"""reconstruction of image"""

for S in range(1,5):
    Sig=np.eye(S)*s[:S]
    mat(Sig)
    U_reduced=U[:,:S]
    V_reduced=V[:S,:]
    
    
    M_reconstruct=U_reduced*Sig*V_reduced
    
    plt.subplot(1,5,S+1,)
    plt.imshow(M_reconstruct)
  
    plt.title("S = %d"%S)
    plt.show()
```
