```python
import numpy as np

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

plt.subplot(3,2,1) 
x1=np.array([1,2,3,1,5,6,5,5,6,7,8,9,7,9])
x2=np.array([1,3,2,2,8,6,7,6,7,1,2,1,1,3])

X=np.array([x1,x2]).T
X.shape
plt.xlim([0,10])
plt.ylim([0,10])

colors=['b','g','r','m','y','k','c','b']
markers=['o','s','D','v','^','p','*','+']
tests=[2,3,4,5,8]

subplot_counter=1
plt.tight_layout()

for  t in  tests:
    subplot_counter+=1
    plt.subplot(3,2,subplot_counter)
    model=KMeans(n_clusters=t).fit(X)
    for i,l in enumerate(model.labels_):
        plt.plot(x1[i],x2[i],color=colors[l],marker=markers[l],ls='None')
    plt.title('K=%s,coeff=%.3f'%(t,silhouette_score(X,model.labels_,metric='euclidean')))
plt.show()
```
