```python

import skimage.io as io #image reading


mandrill=io.imread('/home/jimbrootan/Desktop/ML_data/mandrill.png') #to read image_file using skimage.io
print(mandrill.shape) #identify the shape / pixel resolution of the image
plt.imshow(mandrill)
plt.show()

```

```python
"""conversion of image from RGB to greyscale"""

from skimage.color import rgb2gray
from skimage.exposure import equalize_hist

mandrill=equalize_hist(rgb2gray(mandrill))
plt.imshow(mandrill)#grey_SCale image
mandrill.shape
```

```python
"""CORNER FINDING using CORNER HARRIS AND CORNER PEAKS"""

from skimage.feature import corner_harris,corner_peaks #corner and edge identification

corners=corner_peaks(corner_harris(mandrill),min_distance=2)
corners.shape
```

```python
"""CORNER_DETECTION
1.zipping===>compression, joining"""
import matplotlib.pyplot as plt
def corner_detection(corners,image):
    plt.figure()
    plt.gray()
    plt.imshow(image)
    
    y_corner,X_corner=zip(*corners)
    print(len(y_corner))
    plt.plot(X_corner,y_corner,'or')
    plt.xlim(0,image.shape[1])
    plt.ylim(image.shape[0],0)
    plt.show()

corner_detection(corners,mandrill)
```
